```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from PIL import Image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
```

```python
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if image is None:
    print("image not found")
    exit()

_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

num_labels, labels = cv2.connectedComponents(binary_image)

print(f"✅ Number of connected components: {num_labels - 1}")

label_viz = (labels / num_labels * 255).astype(np.uint8)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(binary_image, cmap='gray')
plt.title("Binary image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(label_viz, cmap='nipy_spectral')
plt.title("Labeled connected components")
plt.axis("off")

output_file = "connected_components.png"
plt.savefig(output_file)
plt.close()

display(Image.open(output_file))
```
