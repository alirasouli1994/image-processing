```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
file_names = list(uploaded.keys())
```

```python
def gamma_correction(image,gamma):
    img_norm = image / 255.0
    img_corrected = np.power(img_norm, gamma)
    img_corrected = (img_corrected * 255).astype(np.uint8)
    return img_corrected

image1 = cv2.imread(file_names[0], cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread(file_names[1], cv2.IMREAD_GRAYSCALE)

gamma_values = [0.4, 0.7, 1.0, 1.5, 2.2]

fig, axes = plt.subplots(len(gamma_values), 2, figsize=(10, 15))

for i, gamma in enumerate(gamma_values):
    corrected1 = gamma_correction(image1, gamma)
    corrected2 = gamma_correction(image2, gamma)

    axes[i, 0].imshow(corrected1, cmap='gray')
    axes[i, 0].set_title(f'{file_names[0]} - Gamma: {gamma}')
    axes[i, 0].axis('off')

    axes[i, 1].imshow(corrected2, cmap='gray')
    axes[i, 1].set_title(f'{file_names[1]} - Gamma: {gamma}')
    axes[i, 1].axis('off')

plt.tight_layout()
plt.show()
```
