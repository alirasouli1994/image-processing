```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files
from IPython.display import display
uploaded = files.upload()
```

```python
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if image is None:
    print ("image not found")
    exit()

L_values = [128, 64, 32, 16, 8, 4, 2]

fig, axes = plt.subplots(2, 4, figsize=(12, 8))

for i, L in enumerate(L_values):
    row, col = divmod(i, 4)

    quantized_image = np.floor(image/(256 / L)) * (256 / L)
    quantized_image = quantized_image.astype(np.uint8)

    ax = axes[row, col]
    ax.imshow(quantized_image, cmap='gray')
    ax.set_title(f"L = {L}")
    ax.axis("off")

if len(L_values) < 8:
    fig.delaxes(axes[1, len(L_values) % 4])

output_file = "quantization_results.png"
plt.savefig(output_file)
plt.close(fig)

display(Image.open(output_file))
```
