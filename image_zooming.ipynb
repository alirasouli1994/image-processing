```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from PIL import Image

uploaded = files.upload()
image_path = list(uploaded.keys())[0]
```

```python

image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Image not found.")
    exit()


nearest_resized = cv2.resize(image, None, fx=4, fy=4, interpolation=cv2.INTER_NEAREST)

bilinear_resized = cv2.resize(image, None, fx=4, fy=4, interpolation=cv2.INTER_LINEAR)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(image, cmap='gray')
axes[0].set_title("master image")
axes[0].axis("off")

axes[1].imshow(nearest_resized, cmap='gray')
axes[1].set_title("Nearest Neighbor")
axes[1].axis("off")

axes[2].imshow(bilinear_resized, cmap='gray')
axes[2].set_title("Bilinear Interpolation")
axes[2].axis("off")

plt.savefig("zooming_results.png")
plt.close()

display(Image.open("zooming_results.png"))
```
